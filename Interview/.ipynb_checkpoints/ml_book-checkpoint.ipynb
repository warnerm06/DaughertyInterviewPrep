```python
# create a matrix 
matrix = np.array([[0,2,0,4,5],
                   [0,3,0,0,7],
                   [0,3,0,0,1]])
```

```python
# Create a sparse matrix
from scipy import sparse

matrix_sparse = sparse.csr_matrix(matrix)
print(matrix_sparse)
```

```python
# select from a matrix
print(matrix[1])
print(matrix[1,1])
print(matrix[1][1])
```

```python
# select from a vector
vector = np.array([1,2,3,4,5,6,7])
            
print(vector[:]) # select all elements
print(vector[1:4]) # select a slice
print(vector[3:]) # select from 3 to end 
print(vector[:3]) # select from beginning to 3
print(vector[-1]) # select last element
print(vector[::2]) # get every other element
print(vector[::-1]) # reverse a list
```

```python
print('max: ',np.max(matrix))
print('min: ',np.min(matrix))
print('std: ',np.std(matrix))
print('mean: ',np.mean(matrix))
print('median: ',np.median(matrix))
print('var: ', np.var(matrix))

from scipy import stats # no mode in numpy
print('mode: ',stats.mode(matrix, axis = None))
```

```python
matrix.flatten() # flatten a matrix
```

```python
print(matrix.reshape(3,-1)) # reshape matrix
print(matrix.reshape(-1,1))
```

```python
import pandas as pd

url='https://tinyurl.com/titanic-csv'

dataframe= pd.read_csv(url)
display(dataframe.head())
display(dataframe.describe())
dataframe.shape
```

```python
display(dataframe.iloc[0:3]) # get rows by index
display(dataframe[dataframe['Sex']=='female'].head()) # filter based on single column values
display(dataframe[(dataframe['Sex']=='female') & (dataframe['Age']>=22)].head()) # filter based on multiple conditions
display(dataframe['Sex'].replace('female','f').head())
```

```python
df = pd.DataFrame()
df['apples']= ['Granny-Smith', 'Gala', 'Honey-Crisp', 'Fuji', 'Yellow', 'Yoshi','Yoshi']
df['price_per_lb']= [1.15, 2.25, .99, 2.50,2.99, None, None]
df['vendor']= ['Moe', 'Larry', 'Curly', 'B.Bunny', 'B.Bunny', 'D.Duck', 'D.Duck']
display(df)

new_apple= pd.Series(['Big Red', 1.99, 'B.Bunny'], index= ['apples', 'price_per_lb','vendor'])
df.append(new_apple, ignore_index=True)
```

```python
# rename columns df.rename(columns={'currentName': 'newName'})
df.rename(columns={'apples':'Apples'}) 
```

```python
print('count: ',df['price_per_lb'].count())
print('min: ',df['price_per_lb'].min())
print('max: ',df['price_per_lb'].max())
print('mean: ',df['price_per_lb'].mean())
print('mode: ',df['price_per_lb'].mode())
```

```python
print(df['apples'].unique()) # get unique values 
print(df['apples'].nunique()) # get number of unique values
print()
print(df['apples'].value_counts())
```

```python
# get all null values in a row
df[df['price_per_lb'].isnull()]
```

```python
display(df)
df.drop('vendor', axis =1) # drop column
df.drop(['vendor', 'price_per_lb'], axis=1) # drop multiple columns
```

```python
# delete a row (or rather, create a df without the row in it)
df[df['apples']!='Fuji']
```
